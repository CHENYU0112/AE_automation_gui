```python
# This code demonstrates how to use the Numpy clip function in order to clip a waveform object on defined thresholds
# It also serves as an example how to use any available Numpy funtion on a pVerify Waveform Object
# Lots of useful functions are documented here: http://docs.scipy.org/doc/numpy/reference/routines.math.html

from wavewatson import Waveform
# only needed for sample data generation
import numpy
# make plots appear inline
%matplotlib inline
```

# create a waveform with some synthetic data

```python
# Time vector
t = numpy.linspace(0,1,10000)
# Data vecor
v = numpy.sin(2 * numpy.pi *10 * t)
# create Waveform object
wave  = Waveform(v, t)
# plot the Sample data
wave.plot()
```

```python
# create a clipped version of teh Waveform and plot it
wave.data = wave.data.clip(-0.3, 0.5)
wave.plot()
```
