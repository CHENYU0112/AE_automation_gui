# class Waveform()

```python
%matplotlib inline
import os
from pathlib import Path
import matplotlib.pyplot as plt
from wavewatson import Waveform
# import pverify.postproc as postproc
from scipy import signal

gd = Waveform.load_from_file(filepath=r"demo_signals\gd.npz")
vcc = Waveform.load_from_file(filepath=r"demo_signals\vcc.npz")
vout = Waveform.load_from_file(filepath=r"demo_signals\vout.npz")
```

```python
gd.plot()
```

### Length

Returns the number of data points of the waveform.

```python
gd.__len__()
```

### Addition

Add to the current waveform, another waveform or numeric value.

```python
gd_tmp = gd.copy()
(gd_tmp+gd_tmp).plot()
```

### Subtraction

Subtract a numeric value, or another waveform from the current waveform.

```python
(gd_tmp-gd_tmp).plot()
```

### Multiplication

Produkt of a numeric value, or another waveform and the current one.

```python
(gd_tmp*gd_tmp).plot()
```

### Division

```python
(gd_tmp/2.0).plot()
```

### Absolute

Data vector converted to absolute values.

```python
abs(gd_tmp).plot()
```

### Invert

Invert the data points.

```python
gd.mirror().plot()
```

### Power

Data vector to the power of the parameter.

```python
(gd_tmp**3).plot()
```

### Equal to

Compare two waveforms if they have the same time and data vectors.

```python
gd.__eq__(vcc)
```

### Greater than

```python
gd.__gt__(gd)
```

### Greater than or equal to

```python
try:
    gd.__ge__(vcc)
except Exception as ex:
    print(ex)
```

### Lower than

```python
try:
    gd.__lt__(vcc)
except Exception as ex:
    print(ex)
```

### Lower than or equal to

```python
gd.__le__(gd)
```

### Set item

### Apply window

```python
w = gd.copy()
w.apply_window("hamming")
w.plot()
```

### At time / interpolate

Interpolate the waveform with the time.

```python
at = gd.copy()
at.at_time(10.2)
at.plot()
```

### Copy

Copy the current waveform

```python
x = gd.copy()
x.plot()
```

### Derivate

Returns the nth derivate of the waveform.

```python
x.Math.derivate(2).plot()
```

### Decimate / Down sample

Decimate the current waveform samples.

```python
d = gd.copy()
d.decimate(50).plot()
print(gd.samplerate, d.samplerate)
```

### Extend

Appends another waveform to the end.

```python
e = gd.copy()
e.extend(gd).plot()
```

### FFT

```python
ft = gd.fft()
ft.plot()
```

### Histogram

```python
gd.histogram().plot()
```

### iFFT

```python
ft.ifft().plot()
```

### Insert start

Insert another waveform at the beginning.

```python
i = gd.copy()
i.insert_start(gd).plot()
```

### Load from file / save to file

load or save waveform from/to a file

```python
savepath = Path(os.environ["TEMP"]) / "dump.npz"

vcc.save_to_file(savepath)
print(vcc)
```

```python
restored_wf = Waveform.load_from_file(filepath=savepath)
print(restored_wf)
```

### Mirror

```python
vcc.mirror().plot()
```

### Repeat

Append the waveform to itself.

```python
r = gd.copy()
r.repeat(3).plot()
```

### Slice by indices

Returns the slice from the waveform from start index to end index.

```python
s = gd.copy()
s.slice_by_indices(100, 500).plot()
```

### Slice by time

```python
st = gd.copy()
st.slice_by_time(0.00004, 0.00011).plot()
```

### Split by indices

Split the waveform in pieces at their indices.

```python
si = gd.copy()
a, b, c = si.split_by_indices([100, 500])
b.plot()
```

### Split by times

```python
t = gd.copy()
a, b = t.split_by_times([0.00005])
a.plot()
```

### Time shift

```python
ts = gd.copy()
ts.time_shift(2).plot()
```

### Up sample

```python
us = gd.copy()
us.up_sample_interp(150000)
print(gd.samplerate, us.samplerate)
```
