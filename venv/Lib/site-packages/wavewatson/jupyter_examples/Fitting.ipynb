# Subclass Fitting

```python
%matplotlib inline
import numpy as np
from wavewatson import Waveform

vcc = Waveform.load_from_file(filepath=r"demo_signals\vcc.npz")
vout = Waveform.load_from_file(filepath=r"demo_signals\vout.npz")

t = np.linspace(0, 1, 10000, endpoint=False)
noise = np.random.normal(0, 0.1, len(t))
sine = Waveform(time=t, data=(10 * np.sin(2 * np.pi * 5 * t) * np.linspace(0,1,len(t))[::-1])+noise)
```

```python
print(vcc)
print(vout)
print(sine)

vcc.plot()
vout.plot()
sine.plot()
```

### Linear best fit

Seperates a waveform into several lines of best fit.

```python
fit_vout, gradient_vout = vout.Fitting.segmented_linear_best_fit(max_err=1, min_segment_len=200, window_reduce_factor=0.8)
fit_vout.plot()
gradient_vout.plot()
```

```python
fit_vcc, gradient_vcc = vcc.Fitting.segmented_linear_best_fit(max_err=2, min_segment_len=300, window_reduce_factor=0.8)
fit_vcc.plot()
gradient_vcc.plot()
```

```python
fit_sine, gradient_sine = sine.Fitting.segmented_linear_best_fit(max_err=2, min_segment_len=50, window_reduce_factor=0.5)
fit_sine.plot()
gradient_sine.plot()
```
