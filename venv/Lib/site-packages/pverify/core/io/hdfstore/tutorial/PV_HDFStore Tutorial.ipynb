# Enabling PV_HDFStore output

To enable this output dataformat, call the method **self.StorageOptions.ConfigurePandasHDFStore(True)** in the **Initialization** method of a PyVerify test.

Per default, the test will create a **HDFStore_data.h5** and **HDFStore_wf.h5** file inside the tests result directory.

# Converting existing results to PV_HDFStore

```python
from pverify import convert_to_pandas_hdfstore
convert_to_pandas_hdfstore(
    source=r"C:\temp\PyVerify\results\Dummies\Avenue_DummyTest\results.hdf5",
    dst_data=r"C:\temp\PyVerify\results\Dummies\Avenue_DummyTest\converted_data.h5",
    dst_wf=r"C:\temp\PyVerify\results\Dummies\Avenue_DummyTest\converted_wf.h5",
    complevel=9,
    complib="blosc"
)
```

# Open/Close a PyVerify HDFStore

```python
from pverify import PV_HDFStore, Waveform
```

```python
DATA_FILE = "HDFStore_data.h5"
WF_FILE = "HDFStore_wf.h5"
```

```python
h = PV_HDFStore(DATA_FILE, mode="r")
# h.Close()
```

# General access and display

## Display parameter/output info

```python
h.pv_select_info()
```

## Display all parameters

```python
h.pv_select_index()
```

## Display all scalar outputs

```python
h.pv_select_data()
```

## Display parameters and scalar outputs

```python
h.pv_select_index_and_data()
```

# Filtered access

## Display filtered parameters

```python
h.pv_select_index(where="Frequency == 8. and Offset >= 0.102")
```

## Display filtered scalar outputs

```python
h.pv_select_data(where="maxLoad > 2700")
```

```python
h.pv_select_data(where="maxLoad > 2700", columns=["maxLoad", "energyAvg"])
```

## Display filtered parameters and scalar outputs

```python
h.pv_select_index_and_data(where_index="Frequency == 8. and Offset >= 0.102",
                           where_data="maxLoad > 2700")
```

Drop the parameter columns:

```python
h.pv_select_index_and_data(where_index="Frequency == 8. and Offset >= 0.102",
                           where_data="maxLoad > 2700", 
                           dropindexcols=True)
```

Only retrieve selected columns:

```python
h.pv_select_index_and_data(where_index="Frequency == 8. and Offset >= 0.102",
                           where_data="maxLoad > 2700",
                           columns=["Amplitude", "maxLoad", "energyAvg"])
```

# Exporting

```python
df = h.pv_select_index_and_data()
```

```python
newdf = df[(df["Damping"] == 2.) & (df["Offset"] >= 0.102)]
print(newdf)
```

## CSV

```python
newdf.to_csv(path_or_buf="C:/temp/newdf.csv", sep="\t")
```

## Excel

```python
newdf.to_excel(excel_writer="C:/temp/newdf.xlsx", sheet_name="new_df")
```

# Retrieving waveforms

## Filter waveforms by parameters

Use the argument 'iterator=True' to not load all waveforms into memory

```python
h.pv_select_from_wfstore(wfstorefile=WF_FILE, where="Frequency == 8. and Offset == 0.105", iterator=True)
```

```python
h.pv_select_from_wfstore(wfstorefile=WF_FILE, where="Frequency == 8. and Offset == 0.105", iterator=False)[0].head()
```

```python
for df in h.pv_select_from_wfstore(wfstorefile=WF_FILE, 
                                   where="Frequency == 8. and Offset == 0.105", 
                                   iterator=True, 
                                   columns=["dampedOsc"]):
    df.dampedOsc.plot()
```

## Filter waveforms by scalar outputs

```python
filter_data = h.pv_select_data(where="maxLoad > 2700")
filter_data
```

```python
for df in h.pv_select_from_wfstore(wfstorefile=WF_FILE, 
                                   where=filter_data.index, 
                                   iterator=True, 
                                   columns=["dampedOsc"]):
    df.dampedOsc.plot()
```
